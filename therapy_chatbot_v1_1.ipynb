```
## To use click the play button to the left of this box and scroll to the bottom :)


## prints a leading phrase to introduce a question or request
def question_prompt():
  question_prompts = [
  "May I ask,", 
  "I'm wondering,", 
  "I'm curious,", 
  "Would you do me the kindness of telling me,", 
  "I'd love to know,",
  "I would enjoy knowing,"
  ]

  i = random.randint(0,len(question_prompts)-1)
  q_prompt = question_prompts[i]
  print(q_prompt)

## returns thankful praise for use of the program as the user answers.
def gratitude_for_use():
  grateful_phrases = [
  "Good job answering.",
  "You're doing a wonderful job caring for yourself.",
  "You are improving the lives of those around you.",
  "I appreciate you taking the time to help out this project.",
  "Your answers will help us create more caring conversations\
  \nfor yourself and others! :)",
  "You are doing great!",
  "You are helping us think better. Thank you!",
  "You matter <3",
  "You are perfect exactly as you are.",
  "I love you.",
  "Thank you for helping us out with this project",
  "I love when do such a wonderful job answering!",
  "Thank you! :)",
  "You have worth.",
  "I appreciate you.",
  "Great job!",
  "Aww <3 That's nice. :)"

  ]
  i = random.randint(0,len(grateful_phrases)-1)
  g_phrase = grateful_phrases[i]
  print("\n" + g_phrase + "\n")
  pass


## Introduce the program to the user
def intro():
  print("If there are any questions you would not like to answer, \
  \nplease write \"pass\" and it will move to the next question.\n")


## prompts a question and  
## saves (appends) the question and the answer to the answers file
def add_answer(question):
  answer = input(question + "\n")
  
	# with open('answers.json', 'a') as answers:
  #   json.dump(answers, [question, answer])
  conversation.append((question, answer))
  return answer


## Asks a question from the question list and returns answer
def run_main_question(question_list):
  print("")
  i = random.randint(0, len(question_list)-1)
  question = question_list[i]
  answer = add_answer(question)
  gratitude_for_use()
  return question, answer

## asks the user to ask themselves a follow up question
def run_followup_q(question):
  followup_question = input("\
  \nWhat follow up question would you most like to be asked?\n")
  if followup_question.lower() == 'pass':
    return
  gratitude_for_use()
  print("")
  question_prompt()
  answer = add_answer(followup_question)
  q_followup = (question, followup_question, answer)
  conversation.append(q_followup)
  gratitude_for_use()
  # with open(followup_questions.json, 'a') as followup_questions:
  #   json.dump(q_followup, followup_questions)



def run_new_question():
  print("")
  new_q_prompt = "What question would you like to be asked by me?"
  new_q = add_answer(new_q_prompt)
  if new_q.lower() == 'pass':
    return
  gratitude_for_use()
  print("")
  question_prompt()
  add_answer(new_q)
  gratitude_for_use()

# runs the file, giving instructions, deciding which list to choose, 
# asking question, retrieving followup question, asking follow up question, 
# and storing these together
def run():
  intro()
  question_list = gratitude_question_list
  while True:
    for i in range(3):
      question, answer = run_main_question(question_list)
      if answer.lower() != 'pass':
        run_followup_q(question)
      run_new_question()
      # run_values_question()
      # run_virtue_affirmation()





run()
```
