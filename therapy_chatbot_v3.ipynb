```python

## Introduce the program to the user
def intro():
    print(intro_text)
    # print(intsrtuctions_text)

## prints a leading phrase to introduce a question or request
def question_prompt():
    i = random.randint(0,len(question_prompts)-1)
    q_prompt = question_prompts[i]
    print(q_prompt)
    return q_prompt

## returns thankful praise for use of the program as the user answers.
def gratitude_for_use():
    i = random.randint(0,len(grateful_phrases)-1)
    g_phrase = grateful_phrases[i]
    print("\n" + g_phrase + "\n")
    pass


def run_commands(response):
    if response.lower() == "improvement":
        improvement_prompt()
    elif response.lower() == "comment":
        feedback_prompt()
    elif response.lower() == "new question":
        run_new_question()
    elif response.lower() == "stop":
        stop_program()


## prompts a question and  
## saves (appends) the question and the answer to the answers file
##
## 
def add_answer(question):
    answer = input(question + "\n")

    run_commands(answer)
  
    # with open('answers.json', 'a') as answers:
    #   json.dump(answers, [question, answer])
    conversation.append((question, answer))


  


  return answer


## Asks a question from the question list and returns answer
def run_main_question(question_list):
    print("")
    i = random.randint(0, len(question_list)-1)
    question = question_list[i]
    question_prompt()
    answer = add_answer(question)
    gratitude_for_use()
    return question, answer

## asks the user to ask themselves a follow up question
def run_followup_q(question):
    followup_prompt = ("\
    \nWould you like to be asked a followup question?\
    \nIf so, what would you like to be asked?\n")
    followup_question = add_answer(followup_prompt)
    gratitude_for_use()
    print("")
    question_prompt()
    answer = add_answer(followup_question)
    q_followup = (question, followup_question, answer)
    conversation.append(q_followup)
    gratitude_for_use()
    # with open(followup_questions.json, 'a') as followup_questions:
    #   json.dump(q_followup, followup_questions)


##
def stop_program():
    exit_message = "\
    \n\
    \n\
    Thank you for taking this time to talk with us.\
    \nWe wish you a day full of joy. <3\
    \n\
    \n"
    print(exit_message)
    sys.exit("All done")


##
def run_new_question():
    print("")
    new_q_prompt = "What question would you like to be asked by me?"
    new_q = add_answer(new_q_prompt)
    gratitude_for_use()
    print("")
    if new_q.lower() == 'pass':
        return new_q_prompt, "pass"

    question_prompt()
    answer = add_answer(new_q)
    gratitude_for_use()
    return new_q, answer


##
def improvement_prompt():
    print("")
    improvement_question = "Thank you, love! <3\
    \nHow can we help?"
    improvement= add_answer(improvement_question)
    gratitude_for_use()
    print("")

##
def affirmation_prompt():
    question = "What words would you most like to hear from me right now?"
    answer = add_answer(question)
    print("")
    print(answer)
    gratitude_for_use()

##
def feedback_prompt():
    print("")
    feedback_question = "We so appreciate you giving us feedback! <3\
    \nWhat's on your mind?"

    comment = add_answer(feedback_question)
    gratitude_for_use()
    print("")

# runs the file, giving instructions, deciding which list to choose, 
# asking question, retrieving followup question, asking follow up question, 
# and storing these together
def run(question_list=journal_question_list):
    intro()

    while True:
      
    ## needs fixing because is being returned none type which is not iterable
        question, answer = run_main_question(question_list)
#     if answer.lower() != 'pass':
#       run_followup_q(question)
#     question, answer = run_new_question()
#     if answer.lower() != 'pass':
#       run_followup_q(question)
#     affirmation_prompt()
    
    # run_values_question()
    # run_virtue_affirmation()





```
